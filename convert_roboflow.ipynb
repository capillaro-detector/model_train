```python
from pathlib import Path
from typing import Any, List
```

```python
levels = ('test', 'train', 'valid')
```

```python
def get_files(files_dir: Path, suff: str = 'txt') -> List[Path]:
    return sorted(Path(files_dir).glob(f'*.{suff}'))
```

```python
def get_lines(curent_file):
    with open(curent_file, 'r') as f:
        return f.readlines()

#data
```

```python
def create_data(data: List[Any]):
    file_data = []
    for line in data:    
        coord = line.split()
        new_coord = []
        size = 8
        for element in coord:
            if len(element) < size:
                new_coord.append(int(element))
                continue
            float_el = float(element[:8])
            new_coord.append(float_el)

        new_coord[-1] = f'{new_coord[-1]}'
        file_data.append(' '.join(new_coord))
    return file_data
```

```python
def create_newfile(file_data, new_dir, curent_file):
    new_file = Path(new_dir, curent_file.name)
    with open(new_file, 'w') as f:
        f.writelines(file_data)
```

```python
def convert_txt():
    for level in levels:
        labels_dir = Path(f'eyes-3/{level}/labels')
        new_dir = Path(f'eyes-3/{level}/new_labels')

        file_list = get_files(labels_dir)
        for yolo_file in file_list:
            file_data = get_lines(yolo_file)
            new_data = create_data(file_data)
            create_newfile(new_data, new_dir, yolo_file)
```

```python
convert_txt()
```
